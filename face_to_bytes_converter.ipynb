# Converting faces to bytes

## Importing Variables

```python
import cv2
import os
import matplotlib.pyplot as plt
import json
```

## Creating a dictonary

```python
dict = {}
```

## Function to covert a img to an array

```python
def png_to_byte(path):
    img = cv2.imread(f'faces/{path}', cv2.IMREAD_GRAYSCALE)
    img = cv2.threshold(img, 0, 1, cv2.THRESH_BINARY)
    return img[1].tolist()
```

## Converting all faces

```python
for img in os.listdir('faces'):
    array = img.split('.')[0]
    dict[array] = png_to_byte(img)
```

## Checking if it's works

```python
plt.figure(figsize=(10, 10))
for i, face in enumerate(dict):
    plt.subplot(5, 7, i+1)
    plt.title(face)
    plt.imshow(dict[face], cmap='gray')
    plt.axis('off')
plt.tight_layout()
plt.show()
```

## Exporting as a json file

```python
with open('faces.json', 'r') as file:
    dict = json.load(file)
```
